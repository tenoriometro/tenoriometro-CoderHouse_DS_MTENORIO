```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/tenoriometro/tenoriometro-CoderHouse_DS_MTENORIO/main/Datasets_desafios/Telecom%20Churn%20Rate%20Dataset.csv'
df = pd.read_csv(url, sep=';')

```

```python
url2 = 'https://raw.githubusercontent.com/tenoriometro/tenoriometro-CoderHouse_DS_MTENORIO/main/Datasets_desafios/Telecom%20Company%20Call-Center-Dataset.csv'
df2 = pd.read_csv(url2, sep=';')
```

```python
#DEFI 2 

# --------------------------------------------------------------------- Matplotilb------------------------------------------------------------------
#Importaciones de bibliotecas para fines de visualizacion de datos : Matplotlib
import matplotlib as mpl
import matplotlib.pyplot as plt

# Crea una figura con tres subplots
fig, axs = plt.subplots(1, 3, figsize=(20, 4))

#1 
valores = df['PhoneService'].unique()
conteos = df['PhoneService'].value_counts()
axs[0].pie(conteos, labels=valores, autopct='%1.1f%%')
axs[0].set_title('¿Cuenta Con Servicio Telefonico?')

#2
payment_method_counts = df["PaymentMethod"].value_counts()
axs[1].bar(payment_method_counts.index, payment_method_counts.values)
axs[1].set_title("Cantidad de Clientes por Metodo de Pago")
axs[1].set_xlabel("Método de Pago")
axs[1].set_ylabel("Cantidad de clientes")
axs[1].tick_params(axis='x', labelsize=6)


#3

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

plt.boxplot(df['TotalCharges'].dropna())
plt.title('Conusmo total de clientes')
plt.ylabel('Monto Total')
plt.show()


```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 1er gráfico: evolución en el tiempo de cantidad de llamadas realizadas
df2['Date'] = pd.to_datetime(df2['Date'])
df2_count = df2.groupby('Date').count()

fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_subplot(1, 3, 1)
sns.lineplot(x='Date', y='Call Id', data=df2_count, ax=ax1)
ax1.set_title('Evolución en el tiempo de cantidad de llamadas realizadas')
ax1.set_xlabel('Fecha')
ax1.set_ylabel('Cantidad de registros')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45, ha='right')
fig.autofmt_xdate(rotation=45)

# 2do gráfico: conteo de resoluciones de llamadas
ax2 = fig.add_subplot(1, 3, 2)
sns.countplot(y="Resolved", data=df2, ax=ax2)
ax2.set_title("Resoluciones de llamadas")
ax2.set_xlabel("Cantidad")
ax2.set_ylabel("Estado de resolución")

# 3er gráfico: duración promedio de las llamadas
ax3 = fig.add_subplot(1, 3, 3)
sns.histplot(data=df2, x="AvgTalkDuration", kde=True, color="blue", bins=50, binwidth=5, ax=ax3)
ax3.set_title("Duración promedio de las llamadas")
ax3.set_xlabel("Duración promedio (segundos)")
ax3.set_ylabel("Frecuencia")

plt.show()
```
